Задача № 121

Построить пример числовой последовательности, имеющей в качестве своих частичных пределов данные числа $a_1, a_2, ..., a_p$

```python
import numpy as np
import matplotlib.pyplot as plt

# Заданные частичные пределы
partial_limits = [0, 1, -1, 0.5, -0.5]

# Длина последовательности
n = 100

# Функция для генерации последовательности
def generate_sequence(partial_limits, n):
    p = len(partial_limits)
    sequence = []
    for i in range(n):
        # Чередуемся между частичными пределами
        a_p = partial_limits[i % p]
        # Добавляем уменьшающийся шум, чтобы последовательность приближалась к пределам
        sequence.append(a_p + np.random.normal(scale=0.1 / (i % p + 1)))
    return sequence

# Генерируем последовательность
sequence = generate_sequence(partial_limits, n)

# Выводим последовательность на графике
plt.figure(figsize=(10, 5))
plt.plot(sequence, marker='o', linestyle='-', color='b', label='Последовательность')
plt.axhline(y=0, color='r', linestyle='--', label='Частичные пределы')
for limit in partial_limits:
    plt.axhline(y=limit, color='r', linestyle='--')
plt.title('Числовая последовательность с заданными частичными пределами')
plt.xlabel('Индекс')
plt.ylabel('Значение')
plt.xticks(np.arange(0, n, step=10))
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Согласно закону Ван-дер-Ваальса объем $v$ реального газа и его давление $p$ при постоянной температуре свзаны соотношением:

$(p+\frac{a}{v^2})(v-b)=c$

Построить график функции $p=p(v)$, если $a=2$, $b=0.1$, $c=10$

Решение:

выразим функцию $p=p(v)$

$p = \frac{c}{v-b} - \frac{a}{v^2}$

```python
def calc_van_der_vaals_pressure(v, a, b, c):
    return (c / (v - b)) - (a / (v**2))

a = 2
b = 0.1
c = 10

v_start = 0.01
v_end = 1
v_steps_num = 1000

v = np.linspace(v_start, v_end, v_steps_num)
p = calc_van_der_vaals_pressure(v, a, b, c)

plt.plot(v, p)
plt.show()
```

Задача 3000

Найти сумму ряда

$\sum_{n=2}^{\infty}\frac{(-1)^n}{n^2 + n -2}$

```python
n = 2

summ = 0
treshold = 1e-9

# функция подсчета следующего слагаемого
def calc_next_term(x: int) -> float:
    return (-1)**x / (x**2 + x - 2)

# добавляем слагаемое к сумме пока величина слагаемого больше граничного значения,
# это означается, что мы следить за изменением суммы, когда сумма начинает слабо меняться,
# считаем, что достигли предела
while abs(calc_next_term(n)) >= treshold:
    summ += calc_next_term(n)
    n += 1

print(f'Сумма ряда: {summ}, количество слагаемых: {n-1}')
```

```python
# построим график сходимости

tresholds = [10**(-i) for i in range(6, 16)]

ns = []
summs = []

for treshold in tqdm(tresholds):
    n = 2
    summ = 0
    while abs(calc_next_term(n)) >= treshold:
        summ += calc_next_term(n)
        n += 1

    ns.append(n-1)
    summs.append(summ)

fig, axs = plt.subplots(nrows= 1, ncols=3, figsize=(15, 5))

axs[0].plot(tresholds, ns)
axs[1].plot(tresholds, summs)
axs[2].plot(ns, summs)
```

```python
# другое решение с применением встроенной библиотеки
import sympy as sp

# Определение переменной
n = sp.symbols('n')

# Определение общего члена ряда
term = (-1)**n / (n**2 + n - 2)

# Разложение знаменателя на множители: n^2 + n - 2 = (n + 2)(n - 1)
factored_term = term.simplify()

# Вывод разложения
print(f'Разложение общего члена ряда: {factored_term}')

# Определение частичной суммы ряда
partial_sum = sp.Sum(factored_term, (n, 2, sp.oo))

# Численное вычисление частичной суммы ряда
approx_sum = partial_sum.evalf()

# Вывод результата
print(f'Приближённая сумма ряда: {approx_sum}')
```

Найти сумму ряда  $\sum_{n=1}^{\infty}\frac{P(n)}{n!}x^n$

```python
m = 3

def polynomial(symbols: list, x: float) -> float:
    return sum([symbols[i] * x**i for i in range(len(symbols))])

# Определение переменных
x, n = sp.symbols('x n')

symbols = [sp.symbols(f'a_{i}') for i in range(m+1)]

term = polynomial(symbols, x) / sp.factorial(n) * x**n

# Определение суммы ряда
series_sum = sp.Sum(term, (n, 1, sp.oo)).doit()
```

Найти ранг матрицы при различных значениях $\lambda$


$[7-\lambda, -12, 6,$ <br>
$10, -19-\lambda, 10,$<br>
$12, -24, 13 - \lambda]$

```python
# создаем переменную lambda
lambda_var = sp.symbols('lambda')

# создаем матрицу
matrix = sp.Matrix([[7 - lambda_var, -12, 6],
               [10, -19 - lambda_var, 10],
               [12, -24, 13 - lambda_var]])

# Функция для вычисления ранга матрицы при данном значении lambda
def compute_rank(matrix, lambda_value):
    matrix_subs = matrix.subs(lambda_var, lambda_value)
    return matrix_subs.rank()

# Пример вычисления ранга для различных значений lambda
lambda_values = [0, 1, -1, 2, -2, 17, -17]
ranks = {value: compute_rank(matrix, value) for value in lambda_values}

# Вывод результатов
for lambda_value, rank in ranks.items():
    print(f'Ранг матрицы при λ = {lambda_value}: {rank}')
```

```python
# Определение символической переменной
lambda_var = sp.symbols('lambda')

# Определение переменных системы
x1, x2, x3, x4 = sp.symbols('x1 x2 x3 x4')

# Определение матрицы коэффициентов и вектора правых частей
A = sp.Matrix([
    [2, -1, 3, 4],
    [4, -2, 5, 6],
    [6, -3, 7, 8],
    [lambda_var, -4, 9, 10]
])

b = sp.Matrix([5, 7, 9, 11])

# Создание расширенной матрицы
augmented_matrix = A.row_join(b)

lambda_value = 3  # 4

matrix_subs = augmented_matrix.subs(lambda_var, lambda_value)
rank = matrix_subs.rank()
print(f'Ранг расширенной матрицы при λ = {lambda_value}: {rank}')
if rank < 4:
    print(f'Система имеет бесконечно много решений при λ = {lambda_value}')
    solutions = sp.linsolve((A, b), x1, x2, x3, x4)
    print(f'Общее решение системы: {solutions}')
else:
    # Найдем общее решение при конкретном λ
    specific_solutions = sp.linsolve((A.subs(lambda_var, lambda_value), b))
    print(f'Решение системы при λ = {lambda_value}: {specific_solutions}')
```

Найти триганометрическую форму числа:

$1 + i\sqrt{3}$


Полярная форма комплексного числа:
$z=a+bi$ записывается как $z=r(cos(\theta)+isin(\theta))$, где:

$r=a2+b2$ — модуль числа

$θ=arg(z)=tg^{−1}(a/b)$ — аргумент числа

```python
# Определение комплексного числа
z = 1 + sp.I * sp.sqrt(3)

# Нахождение модуля
r = sp.Abs(z)

# Нахождение аргумента
theta = sp.arg(z)

# Тригонометрическая форма
trig_form = r * (sp.cos(theta) + sp.I * sp.sin(theta))

# Вывод результатов
print(f'Комплексное число: {z}')
print(f'Модуль: {r}')
print(f'Аргумент: {theta}')
# print(f'Тригонометрическая форма: {trig_form}')
print(f'Тригонометрическая форма: {r}*(cos({theta}) + i*sin({theta}))')
```
